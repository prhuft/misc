```python
%matplotlib notebook
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
```

```python
iters = 40
# ypts = np.empty(num);

# y0 = .5 # initial population, between 0,1
# y = y0
a = .5
# for i in range(0,num):
#     ypts[i]= a*y*(1-y)
#     y = ypts[i]
    
def population(rate,num,s0,fp=None):
    """ sn = rate*sm(1-sm)
        rate in (0,4)
        num is iterations
        s0 is initial population, in (0,1)
    """
    y = s0
    ynew = 0
    ypts = np.empty(num)
    for i in range(0,num):
        ypts[i]= rate*y*(1-y)
        y = ypts[i]
        
#     if fp is not None:
    # need to find the fixed points of the equation....
    # for populations that have oscillatory or sawtooth like behavior,
    # the points it oscillates between are the FPs.
    
    return ypts


fig = plt.figure()
ax = fig.add_subplot(111)
# ax.set_ylim(0,1)
ax.set_xlim(0,num)
# ax.set_aspect(aspect='equal')
ax.set_title("Population")
fig.add_axes(ax)

# ax.plot(ypts)

rates = np.linspace(0.1,3.9,num=10)

# generate population growth vs "time" (iteration) for different rates
for rate in rates:
    pts = population(rate,iters,a)
    ax.plot(pts,label="rate=%f"%rate)
#     ax.legend([rate]) # not sure how to add to the legend each iteration

plt.show()
```
