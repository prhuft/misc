```python
X = [2,3,4]
Y = [5,5,5]
Y2 = [0,0,0]
X3=[3]
xptsU = np.array([(.1*x) for x in range(0,300)])
yptsU = np.array([(x-4)*(x-4) for x in xptsU])
xptsH = [0,0.25,1,1.75,2,2.01,0,-2.01,-2,-1.75,-1,-0.25,0]
yptsH = [0,2,3,2,0,-1,-4,-1,0,2,3,2,0]

fig = plt.figure()
gs = fig.add_gridspec(2, 2)
ax1 = fig.add_subplot(gs[0, 0],xlim=(1,5),ylim=(-1,6))
ax2 = fig.add_subplot(gs[0, 1],xlim=(-3,3),ylim=(-4,4))
# spans two rows:
ax3 = fig.add_subplot(gs[1, 1],xlim=(1,7),ylim=(0,10))

ax1.plot(X,Y,linestyle='-', c='m')
ax1.plot(X,Y2,c='m')
ax1.plot([3,3,3,3,3,3],[0,1,2,3,4,5],c='m')
ax2.plot(xptsH,yptsH,c='r')
ax3.plot(xptsU,yptsU)
```
